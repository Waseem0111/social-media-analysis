# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

# Generate Random Social Media Data

```python
# Define categories
categories = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health']

# Number of records
n = 500

# Generate sample usernames
usernames = [f'user_{i}' for i in range(1, 101)]  # 100 different users

# Generate data
data = {
    'Date': pd.date_range('2021-01-01', periods=n),
    'Category': [random.choice(categories) for _ in range(n)],
    'Likes': np.random.randint(0, 10000, size=n),
    'Shares': np.random.randint(0, 5000, size=n),
    'Content': ['Sample post content ' + str(i) for i in range(n)],
    'Username': [random.choice(usernames) for _ in range(n)]
}
```

# Load Data into DataFrame & Explore

```python
df = pd.DataFrame(data)

# print first few rows
print(df.head())

# DataFrame into
print(df.info())

# Descriptive stats
print(df.describe())

# count of each category
print(df['Category'].value_counts())
```

# Clean the Data

```python
# Remove nulls
df = df.dropna()

# Remove duplicates (make sure to CALL the function with parentheses)
df = df.drop_duplicates()

# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Convert 'Likes' to int
df['Likes'] = df['Likes'].astype(int)
```

# Visualize & Analyze

```python
# Histogram of likes
sns.histplot(df['Likes'], bins=30)
plt.title('Distribution of Likes')
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.show()

# Boxplot: Likes by Category
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Likes', data=df)
plt.title('Likes be Category')
plt.xticks(rotation=45)
plt.show()

# Mean Likes overall
print(f"Average Likes: {df['Likes'].mean():.2f}")

# Mean Likes by Category
print(df.groupby('Category')['Likes'].mean())
```
